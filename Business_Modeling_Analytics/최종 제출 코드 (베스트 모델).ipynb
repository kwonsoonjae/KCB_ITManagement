# Click_Streams 전처리

```python
# Clickstream Data
click = pd.read_csv('test_clickstreams.tab', sep='\t', encoding='cp949')
```

```python
#<접속 카테고리 분석>

# 유저별 BACT(대분류)별 사용 시간 비율

# 유저별 BACT_NM카테고리별 총 사용 시간 그룹핑
sum_catag = click['ST_TIME'].groupby([click['CUS_ID'], click['BACT_NM']]).sum()
# 유저별 총 사용 시간 
cus_time_sum =  click['ST_TIME'].groupby(click['CUS_ID']).sum()
# 유저별 BACK_NM 카테고리별 총 체류시간을 해당 유저의 총 인터넷 체류시간에 대한 비율로 변환
cus_bact_time_percent = sum_catag / cus_time_sum
# unstack 및 NAN을 0으로 변환
cus_bact_time_percent_unstack = cus_bact_time_percent.unstack()
cus_bact_time_percent_fin = cus_bact_time_percent_unstack.fillna(0)
# 결과: cus별 BACT_NM별 ST_TIME 합계의 RATIO 
cus_bact_time_percent_fin
```

```python
#<시간 분석> : 유저별 시간대별 총 체류 시간

click_c = click[['CUS_ID', 'TIME_ID', 'ST_TIME', 'BACT_NM']]
# 뒷 두자리 시간만 추출하기 위해, TIME_ID를 숫자->문자열로 변경
click_c['TIME_ID'] = click_c['TIME_ID'].astype(str)
# TIME_ID 뒷 두자리만 추출하여 TIME_ID_slice 열로 생성
click_c['TIME_ID_slice'] = click_c['TIME_ID'].str.slice(start=8, stop=11)
# TIME_ID_slice를 문자열 -> 숫자열로 변환한 신규 데이터프레임 click_c_new 생성
click_c_new = click_c.astype({'CUS_ID':int, 'TIME_ID':object, 'ST_TIME':float, 'BACT_NM':object, 'TIME_ID_slice':int})
# 24시간을 4가지 시간대로 구분: 1(새벽; 1~6시), 2(오전; 7~12시 ), 3(오후; 13시~18시), 4(밤, 19시~24시)
conditionlist = [
    (click_c_new['TIME_ID_slice'] >= 19) ,
    (click_c_new['TIME_ID_slice'] >= 13 ) & (click_c_new['TIME_ID_slice'] < 19),
    (click_c_new['TIME_ID_slice'] >= 7 ) & (click_c_new['TIME_ID_slice'] < 13),
    (click_c_new['TIME_ID_slice'] < 7)]
choicelist = [4, 3, 2, 1]
click_c_new['TIME_GROUP'] = np.select(conditionlist, choicelist, default='Not Specified')
# 유저에 따른 시간대별 총 체류시간으로 데이터 그룹핑 및 재배열
click_c_new_time = click_c_new['ST_TIME'].groupby([click_c_new['CUS_ID'], click_c_new['TIME_GROUP']]).sum()
click_c_new_time_unstack = click_c_new_time.unstack()
click_timegroup_fin = click_c_new_time_unstack.fillna(0)
# 컬럼 이름 변경
click_timegroup_fin.columns  = ["T1", "T2", "T3", "T4"]
# 결과: T1(새벽; 1시~6시), T2(오전; 7시~12시 ), T3(오후; 13시~18시), T4(밤; 19시~24시)
click_timegroup_fin
```

```python
#<유저의 전반적 사용 행태> :  유저 개인별 인터넷 평균 사용 시간

# 개인별(CUS_ID) 사이트 방문 횟수(SITE_CNT)와 체류시간(ST_TIME)의 각 합계를 구함
cusid_usetime= click.groupby(['CUS_ID'])['SITE_CNT','ST_TIME'].sum()
# 데이터 수집기간 동안 개인별 평균 체류 시간 열(USE_TIME_AVG) 생성
cusid_usetime['USE_TIME_AVG'] = cusid_usetime['ST_TIME']/cusid_usetime['SITE_CNT']
# CUS_ID와 USE_TIME_AVG만 남김
cusid_usetime = cusid_usetime.drop(['SITE_CNT', 'ST_TIME'], axis=1)
# 결과
cusid_usetime
```

```python
# <최종 생성 변수>

#최종결과 1. 카테고리(BACT) & 시간대 & 유저 전반적 행태를 Profile과 병합

# 시간대 & 유저 전반적 행태 병합
time_final = pd.merge(click_timegroup_fin, cusid_usetime , left_index=True, right_index=True, how='outer')
# 카테고리(BACT) & 시간대 & 유저 전반적 행태 병합
click_final = pd.merge(time_final, cus_bact_time_percent_fin, left_index=True, right_index=True, how='outer')
# 결과
click_final
```

# Search_Keywords 전처리

```python
search = pd.read_csv('test_searchkeywords.tab', sep='\t', encoding='cp949')
```

```python
# 검색어에서 명사 추출 후 유저별로 정리

new_list = []
for text in search['QRY_STR']: 
    if '&query=' in text: new_list.append(text[re.search('&query=', text).end():])
    elif '&sm=top' in text: new_list.append(text[:re.search('&sm=top', text).start()])
    else: new_list.append(text)
search['str'] = new_list

Okt = Okt()
search['명사'] =search['str'].apply(Okt.nouns)
search['명사2'] = search['명사'] * search['QRY_CNT'].astype(int)
search2 = search.groupby('CUS_ID')['명사2'].sum()
```

```python
# LDA를 활용한 토픽 모델링

model = tp.LDAModel(k=30, alpha=0.1, eta=0.01, min_cf=3, tw=tp.TermWeight.IDF)
for i in search2:
    if len(i) == 0: model.add_doc([''])
    model.add_doc(i)
model.train(200)

buckets = [[] for _ in range(model.k)]
for d in model.docs:
    for i in range(model.k):
        buckets[i].append(d.get_topic_dist()[i])

probs = pd.DataFrame(search2)
for i in range(model.k):
    probs[i] = buckets[i]
probs = probs.drop('명사2', axis=1)
```

```python
# Click 데이터와 Search 데이터 병합

click = click_final
search = probs

test = pd.merge(click, search, left_index=True, right_index=True, how='outer')
test = test.fillna(0)

test
```
